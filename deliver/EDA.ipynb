## Segmentation

The interaction between categorical and numeric variables.

```python
# tx_price vs property_type
sns.boxplot(x='tx_price', y='property_type', data=df, )
```

```python
# segment by property type and find the mean for each class
df.groupby('property_type').mean()
```

```python
# segment by sqft and property_type
sns.boxplot(x='sqft', y='property_type', data=df)
```

```python
# find the mean and std for each property type
df.groupby('property_type').agg(['mean', 'std'])
```

## Correlations

```python
correlations = df.corr()

sns.heatmap(correlations)
```

```python
# change background to white

sns.set_style('white')

plt.figure(figsize=(10,8))

#sns.heatmap(correlations)

sns.heatmap(correlations * 100, annot=True, fmt = '.0f')
```

```python
# mask the upper triangle
correlations.shape
#np.zeros_like
```

```python
# get the array of booleans with same size of correlations
mask = np.zeros_like(correlations, dtype=bool)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(10,8))

# display heat map along with mask
sns.heatmap(correlations * 100, annot=True, fmt='.0f', mask=mask)
```
